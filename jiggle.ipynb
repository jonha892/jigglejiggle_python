````python
def genres_from_artists(artist_names, artists_df):
    genres = set()
    for name in artist_names:
        if name in artists_df.index:
            genres.update(artists_df.loc[name]["genres"])
    return list(genres)

tracks_df["genres"] = tracks_df["artists"].map(lambda names: genres_from_artists(names, artists_df))
````

````python
from tqdm.auto import tqdm

def get_genre_embedding(genre, openai, model="text-embedding-3-small"):
    response = openai.embeddings.create(input=genre, model=model)
    return response.data[0].embedding, response.usage.total_tokens

embedding_tokens = 0
genre_embeddings = {}
for genre in tqdm(query_genres):
    genre_embeddings[genre], tokens_costs = get_genre_embedding(genre, openai)
    embedding_tokens += tokens_costs
print(f'tokens costs: {embedding_tokens} for {len(query_genres)} genres')
````

````python
from sklearn.cluster import KMeans
import numpy as np

X = np.array(list(genre_embeddings.values()))
````

````python
kmeans = KMeans(n_clusters=7, random_state=0).fit(X)
````

````python
kmeans.cluster_centers_
````

````python
# assign genres to clusters

genre_clusters = {}
for genre, embedding in genre_embeddings.items():
    cluster = kmeans.predict([embedding])[0]
    cluster = f"cluster_{cluster}"
    if cluster not in genre_clusters:
        genre_clusters[cluster] = []
    genre_clusters[cluster].append(genre)
````

````python
# save to file
with open('data/result-genre-clusters.json', 'w') as f:
    json.dump(genre_clusters, f, indent=4)
````
